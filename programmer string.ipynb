```python
def programmerStrings(s):
    # Write your code here
    template = "programmer"
    n = len(template)
    for i in range(n, len(s)-n+1):
        if s[:i].count("p") >= 1 and s[:i].count("r") >= 3 and s[:i].count("o") >= 1 and s[:i].count("g") >= 1 and s[:i].count("a") >= 1 and s[:i].count("m") >= 2 and s[:i].count("e") >= 1:
            rightmost = i
            break
    for i in range(len(s)-n, rightmost-1, -1):
        if s[i:].count("p") >= 1 and s[i:].count("r") >= 3 and s[i:].count("o") >= 1 and s[i:].count("g") >= 1 and s[i:].count("a") >= 1 and s[i:].count("m") >= 2 and s[i:].count("e") >= 1:
            leftmost = i
            break
    return leftmost - rightmost
```

```python
s = "programmerxxxprozmerqgram"
programmerStrings(s)
```
