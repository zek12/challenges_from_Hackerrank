```python
def getMinimumSum(arr):
    min_sum_found = False
    min_sum = 0
    already_seen = set()

    for i in range(len(arr)-2):
        if (not min_sum_found or (min_sum_found and arr[i] < min_sum)) and not str(arr[i]) in already_seen:
            for j in range(i+1, len(arr)-1):
                if (not min_sum_found or (min_sum_found and arr[i]+arr[j] < min_sum)) and not str(arr[i])+str(arr[j]) in already_seen:
                    if not arr[i] >= arr[j]:
                        for k in range(j+1, len(arr)): 
                            if (not min_sum_found or (min_sum_found and arr[i]+arr[j]+arr[k] < min_sum)) and not str(arr[i])+str(arr[j])+str(arr[k]) in already_seen:   
                                if not arr[k] >= arr[j]:
                                    cur_sum = arr[i] + arr[j] + arr[k]
                                    if not min_sum_found:
                                        min_sum_found = True
                                        min_sum = cur_sum
                                    elif cur_sum < min_sum:
                                        min_sum = cur_sum
                            already_seen.add(str(arr[i])+str(arr[j])+str(k))
                    already_seen.add(str(arr[i])+str(arr[j]))
            already_seen.add(str(arr[i]))

    if not min_sum_found:
        return -1
    return min_sum
```

```python
n = 7
arr = [3,4,5,1,2,3,1]
getMinimumSum(arr)
```
