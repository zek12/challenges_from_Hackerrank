```python
def raisingPower(arr):
    current_max = 0
    current_i = 0
    for i in range(len(arr)-1):
        temp = pow(arr[i], arr[i+1]) % (pow(10, 9) + 7)
        if temp > current_max:
            current_max = temp
            current_i = i
    return(current_i + 1)
```

```python
arr = [3,5,4,5,2,10]
raisingPower(arr)
```
