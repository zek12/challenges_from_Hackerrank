```python
def minimumBribes(q):
    bribe = 0
    for i in range(len(q) - 1, -1, -1):
        if q[i] - (i + 1) > 2:
            print("Too chaotic")
            return
        for j in range(max(0, q[i] - 2), i):
            print("comparing pos " + str(j) + " with " + str(i))
            if q[j] > q[i]:
                bribe += 1
    print(bribe)
```

```python
q = [1, 2, 3, 5, 4, 6, 7, 8]
minimumBribes(q)
```

```python
q = [4,1,2,3]
minimumBribes(q)
```
