```python
def bestTrio(friends_nodes, friends_from, friends_to):
    d = {}
    for i in range(len(friends_from)):
        if not friends_from[i] in d.keys():
            d[friends_from[i]] = []
        if not friends_to[i] in d.keys():
            d[friends_to[i]] = []
        d[friends_from[i]] = d[friends_from[i]] + [friends_to[i]]
        d[friends_to[i]] = d[friends_to[i]] + [friends_from[i]]
    
    
```

```python
friends_nodes = 6
friends_from = [1,2,2,3,4,5]
friends_to = [2,4,5,5,5,6]
bestTrio(friends_nodes, friends_from, friends_to)
```
